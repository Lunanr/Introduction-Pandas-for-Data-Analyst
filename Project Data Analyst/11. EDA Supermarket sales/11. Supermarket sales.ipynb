## 9. Display daily sales by Day of The Week

```python
data.columns
```

```python
dw_mapping={
    0:'Mon',
    1:'Tue',
    2:'Wed',
    3:'Thur',
    4:'Fri',
    5:'Sat',
    6:'Sun'
}
```

```python
data['day_of_week']=data['Date'].dt.dayofweek.map(dw_mapping)
```

```python
data['day_of_week'].value_counts().plot(kind='bar')
```

`Daily sales highest on Saturday because there because weeked, the second is Tuesday, and the lowest on Monday because people starting working`

## 10. What will be the highest months for sales

```python
month_mapping = {
    1:'Jan',
    2:'Feb',
    3:'March'
}
```
