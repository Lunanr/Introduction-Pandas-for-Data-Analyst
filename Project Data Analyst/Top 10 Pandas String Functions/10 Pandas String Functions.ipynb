```python
import pandas as pd

dict1 = {'Name': ['Priyang Bhatt', 'Rashmin Panchal', 'Anil Rana', 'Rahul Patel'],
        'City': [' New York ', ' Los Angeles ', ' Chicago', ' Houston'],
        'State': ['NY', 'CA', 'IL', 'TX']}
data = pd.DataFrame(dict1)
```

```python
data
```

# 1. str.lower()

```python
data['name_lower']=data['Name'].str.lower()
#to make a lower string
```

```python
data
```

```python
data['name_upper']=data['Name'].str.upper()
#make a capital string
```

```python
data
```

# 2. str.len()

```python
data['Name_len']=data['Name'].str.len()
#count length of the string
```

```python
data
```

# 3. str.strip()

```python
data['City'][0]
```

```python
data['City_strip']=data['City'].str.strip()
#to delete a white space for the string column
```

```python
data['City_strip'][0]
```

```python
data['city_lstrip']=data['City'].str.lstrip()
#to delete a left white space for string
```

```python
data['city_lstrip'][0]
```

```python
data['city_rstrip']=data['City'].str.rstrip()
#to delete a right white space for string
```

```python
data['city_rstrip'][0]
```
